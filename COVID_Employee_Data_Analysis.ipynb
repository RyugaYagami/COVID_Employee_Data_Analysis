# Project Title: COVID Employee Data Analysis (Pandas Project)

```python
import pandas  as pd
import matplotlib.pyplot as plt


df  =pd.read_csv('COVID_Employee.csv')  # Add  dataset


print("COVID-19 Employee Data \n")  

print(df)

print("____" * 30)

print("\n Department-wise  Employees ValueCount :")

# The Deparment wise  Employees  Vlaue Count result : 
Department_value =df['Department'].value_counts()

print("Department  Total Value count is : ")

print(Department_value)

print("____" * 30)



# Average Salary for Per Department: 

print("The Average Salary For Per Department\n")

Department_salary_ave =df.groupby('Department')['Salary'].mean()

print("The Average Salary For Per Department result: \n", Department_salary_ave)

print('___' * 30)


# Work Model Distribution:

print("The Work Model Distribution:")

work_mode=df['Work_Mode'].value_counts()

print("The Work Model Distribution Value count rusult : \n", work_mode)

print("___" * 30)



# The  COVID_Positive vs Attendance_Days Average: 

print("The COVID_Positive vs Attendance_Days Average:")


Covid_ave =df.groupby("COVID_Positive")['Attendance_Days'].mean()

print("The COVID_Positive Attendance_Days Average result:", Covid_ave)


print("___" * 30)


# Hight Salary Employees : 


Hight_Salary_employees =df.loc[df['Salary'].idxmax()]
print("The Hight Salary Employees Result is :\n", Hight_Salary_employees)

print('___' * 30)


# The  Visualization optional : 


df =df['Work_Mode'].value_counts()


print("The  Work Model  Value Count  \n: ", df)


ax=df.plot(kind='bar',  color='lightgreen', edgecolor='black')


ax.set_title("Work Mode Distribution (COVID Period)")

ax.set_xlabel("Work Mode")

ax.set_ylabel("Employee Count")

ax.grid(True, linestyle='--', alpha=0.6)

ax.length
plt.show()



```
